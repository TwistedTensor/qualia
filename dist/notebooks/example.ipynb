```python
from collections import defaultdict
```

```python
def map_to_graph(mapper_obj,original_data,functions={}):
    G = defaultdict(list)

    for (i,n) in enumerate(mapper_obj.nodes):
        summary = {
               'level': n.level, 
               'n_members':len(n.points), 
               'avg_filter_val': n.attribute.item()}
        data = original_data[n.points,:]
        for name, function in functions.items():
            summary[name] = function(data)
        G['nodes'].append({'id':i,'summary':summary})
        

    G['links'] = [{'source': e[0], 'target': e[1], 'n_overlap':
                   mapper_obj.simplices[1][e]} for e in
                  mapper_obj.simplices[1].keys()]

    return G
```

```python
g = M.to_d3js_graph()
```

```python
import zmq
context = zmq.Context()
```

```python
socket = context.socket(zmq.PAIR)
socket.connect('tcp://127.0.0.1:5560')
```

```python
cgm = {
    'type':'ws/CHANGE_GRAPH',
    'payload':g
}
socket.send_json(cgm)
```
